<h3>1. Simple x-y plots</h3>

```python
import matplotlib.pyplot as plt

# Plot (0,1), (1,2), (2,3)
plt.plot([1,2,3])
plt.ylabel("Y Axis")
plt.show()
```

```python
# Once show() was called, the plot buffer is deleted
plt.xlabel("X Axis")
plt.show()
```

```python
# Points instead of lines with "ro" (red dots)
plt.plot([1,2,3,4], [1,4,9,16], "ro")
plt.axis([0, 6, 0, 20])
plt.show()
```

<h3>2. Multiple curve static x-y plotting</h3>

```python
import numpy

t = numpy.arange(0., 5., 0.2)

# Plot x, x^2 and x^3 from 0 to 5 in 0.2-steps
# Range,Formula,Style,  Range,Formula,Style,...
plt.plot(t,t,'ro',  t,t**2,'bs',  t,t**3,'g')
plt.show()
```

<h3>3. Multiple plot x-y plotting</h3>
A more complex module can be found under https://matplotlib.org/users/gridspec.html

```python
def f(t):
    return numpy.exp(-t) * numpy.cos(2*numpy.pi*t)

t1 = numpy.arange(0.0, 5.0, 0.1)
t2 = numpy.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(2, 1, 1) # Number rows, number lines, figure
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(2, 1, 2) # Number rows, number lines, figure
plt.plot(t2, numpy.cos(2*numpy.pi*t2), 'r--')
plt.yscale('logit') # Logarithmic scale
plt.show()
```

<h3>4. Histograms, text annotation, and saving to a vector file (.svg)</h3>
Vector files can be zoomed endlessly without quality loss.

```python
mu, sigma = 0, 1
x = mu + sigma * numpy.random.randn(10000)

n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('numbes')
plt.ylabel('other numbers')
plt.title('Histogram of numbers')
plt.text(60, .025, 'μ=100, σ=15')
plt.axis([-5, 5, 0, 0.4])
plt.grid(True)
plt.savefig("./Matplotlib/histogram.svg")
plt.show()
```

<h3>5. Legends</h3>

```python
# Directly derived from the User's Guide

line1, = plt.plot([1,2,3], label="Line 1", linestyle='--')
line2, = plt.plot([3,2,1], label="Line 2", linewidth=4)

# Create a legend for the first line.
first_legend = plt.legend(handles=[line1], loc=1)

# Add the legend manually to the current Axes.
ax = plt.gca().add_artist(first_legend)

# Create another legend for the second line.
plt.legend(handles=[line2], loc=4)

plt.show()
```

<h3>7. 3D scatter plots</h3>

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=[1,2,3], ys=[1,2,3], zs=[1,2,3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
